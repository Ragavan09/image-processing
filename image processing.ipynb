```python
virat = cv2.imread('virat.jpeg')
virat_gray = cv2.imread('virat.jpeg',0)
virat_rgb = cv2.cvtColor(virat,cv2.COLOR_BGR2RGB)
virat_hsv = cv2.cvtColor(virat,cv2.COLOR_BGR2HSV)
images = [virat,virat_rgb,virat_gray,virat_hsv]
title = ['BGR IMAGE','RGB IMAGE','Grayscale','virat_hsv']
plt.figure(figsize = (15,8))
for i in range (len(images)):
    plt.subplot(2,2,i+1)
    plt.title(title[i],fontsize = 15)
    if title[i]=='grayscale':
        plt.imshow(images[i],cmap='gray')
    else:
        plt.imshow(images[i])
plt.show()
```

##Drawing Shapes on image

```python
img = np.zeros((200,200,3),np.uint8)
img[:] = (55,81,131)
cv2.line(img,(50,150),(150,50),(255,255,0),3)
cv2.line(img,(50,50),(150,150),(255,255,0),3)
plt.figure(figsize=(6,6))
plt.imshow(img)
plt.show()
```

```python
img = np.zeros((200,200,3),np.uint8)
img[:] = (55,81,131)
cv2.rectangle(img,(50,150),(150,50),(255,255,0),-1)
plt.figure(figsize=(6,6))
plt.imshow(img)
plt.show()
```

```python
img = np.zeros((200,200,3),np.uint8)
img[:] = (55,81,131)
cv2.circle(img,(100,100),50,(255,255,0),4)
cv2.circle(img,(100,100),5,(255,0,0),4)
plt.figure(figsize=(6,6))
plt.imshow(img)
plt.show()
```

## write text on image

```python
img = np.zeros((200,200,3),np.uint8)
img[:] = (55,81,131)
cv2.putText(img,'Ragavan',(50,100),cv2.FONT_HERSHEY_SCRIPT_COMPLEX,1,(255,255,0),2)
plt.figure(figsize=(6,6))
plt.imshow(img)
plt.show()
```
